```python
import pandas as pd
```

```python
# Load the Excel file to inspect sheet names
file_path = '(반출)홈페이지_공시실_상품공시_전체상품_20240913.xlsx'
```

```python
# Load Excel file to inspect sheet names
xls = pd.ExcelFile(file_path)
xls.sheet_names
```

```python
# Load the specified sheets "1. 판매중" and "2. 판매중지"
df_selling = pd.read_excel(file_path, sheet_name='1. 판매중')
df_discontinued = pd.read_excel(file_path, sheet_name='2. 판매중지')

# Preview a few rows to inspect the structure and locate columns for meta data
df_selling.head(), df_discontinued.head()
```

```python
# Adjusting the column selection to correctly match the available data
# We'll use columns C2 to L2 based on the structure previewed earlier

columns_meta_adjusted = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 
                         'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 
                         'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13']

# Filter the metadata from both dataframes
df_selling_filtered = df_selling[columns_meta_adjusted]
df_discontinued_filtered = df_discontinued[columns_meta_adjusted]

# Add a new column to identify the source sheet
df_selling_filtered['Source'] = '판매중'
df_discontinued_filtered['Source'] = '판매중지'

# Merge both sheets into one dataframe
df_combined = pd.concat([df_selling_filtered, df_discontinued_filtered], ignore_index=True)

# Rename the columns according to the metadata headers (META01 to META12)
meta_headers_adjusted = ['META01', 'META02', 'META03', 'META04', 'META05', 'META06', 'META07', 
                         'META08', 'META09', 'META10', 'META11', 'META12', 'Source']
df_combined.columns = meta_headers_adjusted

# Add a "순번" column with sequential numbering starting from 1
df_combined.insert(0, '순번', range(1, len(df_combined) + 1))

# Remove the second row (which contains metadata labels)
df_combined_cleaned = df_combined.drop(0).reset_index(drop=True)

# Since "순번" already exists, we can update it rather than inserting it again
df_combined_cleaned['순번'] = range(1, len(df_combined_cleaned) + 1)

df_combined_cleaned
```

```python
# We will create a new DataFrame where each "순번" will be expanded to multiple rows 
# (one for each of META09, META10, META11, META12), and we will add a "출처" column.

# Define the META columns that will be reshaped
meta_columns = ['META09', 'META10', 'META11', 'META12']

# We will collect the expanded rows in a list to create the new DataFrame
expanded_rows = []

# Iterate through each row in the cleaned dataframe
for index, row in df_combined_cleaned.iterrows():
    for meta_col in meta_columns:
        # Create a new row with the same data as the original row, but with the value from one of the META columns
        new_row = row.copy()
        new_row['출처'] = meta_col
        new_row['파일경로'] = new_row[meta_col]
        expanded_rows.append(new_row)

# Create a new DataFrame from the expanded rows
df_expanded = pd.DataFrame(expanded_rows)

# Drop the original META09, META10, META11, META12 columns as they are now redundant
df_expanded.drop(columns=meta_columns, inplace=True)

df_expanded
```

```python
# Adjust the code to update the column name from '출처' to '상품내역유형'
# and replace the values in this column using the type_mapping

# Define the mapping for '상품내역유형'
type_mapping = {
    'META09': '상품요약서',
    'META10': '사업방법서',
    'META11': '판매약관',
    'META12': '상품특약'
}

# Replace '출처' with '상품내역유형' and apply the type_mapping
df_expanded['상품내역유형'] = df_expanded['출처'].map(type_mapping)

# Drop the old '출처' column as it's no longer needed
df_expanded.drop(columns=['출처'], inplace=True)

df_expanded
```

```python
# Save the merged dataframe to a new Excel file using openpyxl
output_file_path = '통합상품_병합.xlsx'

# Use openpyxl as the engine
with pd.ExcelWriter(output_file_path, engine='openpyxl') as writer:
    df_expanded.to_excel(writer, sheet_name='통합상품', index=False)

output_file_path
```
